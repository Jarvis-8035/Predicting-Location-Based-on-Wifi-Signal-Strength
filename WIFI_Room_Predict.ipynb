```python
X = df[[1,2,3,4,5,6,7]]
Y = df[8]
tX = tdf[[1,2,3,4,5,6,7]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_tr , X_t , Y_tr , Y_t = train_test_split(X,Y,test_size=.30,random_state=42)
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_tr,Y_tr)
```

```python
p_gnb = gnb.predict(X_t)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(Y_t,p_gnb)
```

```python
sol = pd.DataFrame({'Id':tdf[0],'Category':pd.Series(p_gnb)},columns=['Id','Category'])
sol.head()
```
